# Regression Models Draft

## 1

```python
hours_passed = np.random.uniform(7, 10, 8) 
hours_failed = np.random.uniform(1, 4, 12)
```

```python
hours = np.concatenate([hours_passed, hours_failed])
marks = np.concatenate([np.ones_like(hours_passed), np.zeros_like(hours_failed)])
```

```python
np.vstack([hours, marks]).T
```

```python
plt.scatter(hours, marks)
plt.yticks([0, 1], labels=["fail", "pass"])
plt.xlabel("Study time [hours]")
plt.show()
```

## 2

```python
hours = np.random.uniform(1, 10, 20)

def calculate_exam_score(study_hours):
    if study_hours <= 3:
        return 0
    elif study_hours >= 7:
        return 1
    else:
        return int(np.round(np.random.random()))

scores = np.vectorize(calculate_exam_score)(hours)
```

```python
plt.scatter(hours, scores)
plt.yticks([0, 1], labels=["fail", "pass"])
plt.xlabel("Study time [hours]")
plt.show()
```

```python
plt.scatter(hours, scores)
plt.vlines(5, 0, 1, color="red")

plt.yticks([0, 1], labels=["fail", "pass"])
plt.xlabel("Study time [hours]")
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 500)

y1 = sigmoid(x)
y2 = sigmoid(10 * x)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim(-2, 2)
plt.show()
```

```python
def generate_output(x):
    z = sigmoid(x)
    if z < 0.5:
        return 0
    else:
        return 1
```

```python
x = np.arange(-2, 2, 0.4)
np.vectorize(generate_output)(x), x
```

```python
def logistic_function(x, a, b):
    return sigmoid(a * x + b)
```

```python
plt.scatter(hours, scores)

x = np.linspace(-5, 10, 500)
for a, b in zip([1, 2, 3], [3, -8, -10]):
    y = sigmoid(a * x + b)
    plt.plot(x, y)

plt.yticks([0, 1], labels=["fail", "pass"])
plt.xlabel("Study time [hours]")
plt.show()
```

```python
def threshold_logistic_function(x, threshold):
    if x <= threshold:
        return 0
    else:
        return 1
```

```python
a, b = 3, 4
logistic_function_values = logistic_function(hours, a, b)
threshold = b / a
predicted_scores = np.vectorize(threshold_logistic_function)(logistic_function_values, threshold)
```

```python
plt.scatter(hours, scores, alpha = 0.5)
# plt.scatter(hours, logistic_function_values, alpha = 0.5, marker="x")
plt.scatter(hours, predicted_scores, marker="x")
```
