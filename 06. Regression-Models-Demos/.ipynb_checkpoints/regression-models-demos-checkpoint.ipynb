```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.datasets import load_boston
from sklearn.metrics import r2_score
```

# Regression Models Demos

```python
x = np.linspace(-3, 5, 10)
y = 2 * x + 3
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
y_noise = np.random.normal(size = len(y))
y_with_noise = y + y_noise
plt.scatter(x, y_with_noise)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
for y_guess in [3 * x + 8, 4 * x + 3, -2 * x]:
    plt.scatter(x, y_with_noise)
    plt.plot(x, y_guess)
    plt.show()
```

```python
def calculate_loss(x, y, a, b):
    y_predicted = a * x + b
    distances = (y - y_predicted) ** 2
    return np.sum(distances) / len(x)
```

```python
lr = LinearRegression()
lr
```

```python
lr.fit(x.reshape(-1, 1), y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
a = lr.coef_[0]
b = lr.intercept_
y_lr = a * x + b
```

```python
plt.scatter(x, y_with_noise)
plt.plot(x, y_lr)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.scatter(x, y_with_noise)
for y_guess in [3 * x + 8, 4 * x + 3, -2 * x, y_lr]:
    plt.plot(x, y_guess)
plt.show()
```

```python
a_gradient = -2 / len(x) * np.sum(x * (y - (a * x + b)))
b_gradient = -2 / len(y) * np.sum(y - (a * x + b))
```

```python
def perform_gradient_descent(x, y, a, b, learning_rate):
    a_gradient = -2 / len(x) * np.sum(x * (y - (a * x + b)))
    b_gradient = -2 / len(y) * np.sum(y - (a * x + b))
    new_a = a - a_gradient * learning_rate
    new_b = b - b_gradient * learning_rate
    return (new_a, new_b)
```

```python
model_a, model_b = -10, 20 # Start points; can be anywhere
alpha = 0.01 # Learning rate
for step in range(1001):
    model_a, model_b = perform_gradient_descent(x, y, model_a, model_b, alpha)
    if step % 100 == 0:
        error = calculate_loss(x, y, model_a, model_b)
        print("Step {}: a = {}, b = {}, J = {}".format(step, model_a, model_b, error))
print("Final line: {} * x + {}".format(model_a, model_b)) 
```

```python
boston = pd.DataFrame(
    data = load_boston().data,
    columns = load_boston().feature_names,
)
```

```python
boston
```

```python
y = load_boston().target
```

```python
reg = LinearRegression().fit(boston, y)
```

```python
list(zip(boston.columns, reg.coef_))
```

```python
y_pred = boston.apply(lambda row: np.sum(row * reg.coef_) + reg.intercept_, axis = 1)
y_pred
```

```python
r2_score(y_true = y, y_pred = y_pred)
```
