# Regression Models Draft

### Generate exam sample data

```python
EXAM_FAILURE = 3
EXAM_SUCCESS = 5
```

```python
study_hours = np.random.uniform(1, 10, 20)

def calculate_exam_score(study_hours):
    if study_hours <= 3:
        return EXAM_FAILURE
    elif study_hours >= 7:
        return EXAM_SUCCESS
    else:
        return int(EXAM_FAILURE + (EXAM_SUCCESS - EXAM_FAILURE) * np.round(np.random.random()))

exam_scores = np.vectorize(calculate_exam_score)(study_hours)
```

```python
plt.scatter(study_hours, exam_scores)
plt.yticks([EXAM_FAILURE, EXAM_SUCCESS], labels=["Failed exam", "Passed exam"])
plt.xlabel("Study time [hours]")
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 500)
y1 = sigmoid(x)
y2 = sigmoid(10 * x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim(-2, 2)
plt.title("Visualizing two sigmoid functions")
plt.show()
```

```python
def modeling_function(x):
    return EXAM_FAILURE + (EXAM_SUCCESS- EXAM_FAILURE) * sigmoid(x)
```

```python
x = np.linspace(-10, 10, 500)
y1 = modeling_function(x)
y2 = modeling_function(10 * x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.title("Visualizing the scaled modeling function sigmoid functions")
plt.show()
```

```python
def generate_discrete_exam_score(x):
    z = modeling_function(x)
    if z < (EXAM_FAILURE + EXAM_SUCCESS) / 2:
        return EXAM_FAILURE
    else:
        return EXAM_SUCCESS
```

```python
x = np.linspace(-10, 10, 500)
y = modeling_function(x)

x_discrete = np.linspace(-10, 10, 25)
y_discrete = np.vectorize(generate_discrete_exam_score)(x_discrete)

plt.plot(x, y)
plt.scatter(x_discrete, y_discrete)
plt.title("Visualizing conitnuous vs discrete modeling function")
plt.show()
```
