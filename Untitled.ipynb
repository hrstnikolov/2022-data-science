```python
crack = skimage.io.imread('../sample_32_vcc.JPG')
```

```python
def convert_to_grayscale(image):
    red, green, blue = np.split(crack, indices_or_sections=3, axis=2)
    grayscale_image = 0.299 * red + 0.587 * green + 0.114 * blue 
    
    return grayscale_image.astype("uint8")

# crack = crack.mean(axis=2).astype("uint8")
crack = convert_to_grayscale(crack)
```

```python
def threshold_image(image, low, high):
    new_image = image.copy()
    new_image[new_image <= low] = 0
    new_image[new_image >= high] = 255
    
    return new_image

crack_tresholded = threshold_image(crack, 50, 200)
```

```python
plt.imshow(crack.mean(axis=2).astype("uint8"), cmap="gray")
plt.show()

plt.imshow(convert_to_grayscale(crack), cmap="gray")
plt.show()


plt.imshow(threshold_image(crack, 50, 200), cmap="gray")
plt.show()

plt.imshow(threshold_image(crack, 80, 160), cmap="gray")
plt.show()

plt.imshow(threshold_image(crack, 20, 230), cmap="gray")
plt.show()

plt.imshow(threshold_image(crack, 100, 230), cmap="gray")
plt.show()
```
